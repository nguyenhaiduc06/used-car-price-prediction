```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/processed_cars.csv")
df.info()
outliers = df[ df['price'] >= 20000000000].index
df = df.drop(outliers)
```

```python
sns.displot(x='price', data=df)
```

```python
def countplot(x):
  ax = sns.countplot(x=x, data=df, palette="bright")
  ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right",fontsize=8);
def scatterplot(x, y):
  ax = sns.catplot(x=x, y=y, data=df)
  ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right",fontsize=8);
```
