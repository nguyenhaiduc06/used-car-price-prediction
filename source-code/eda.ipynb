```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('./data.csv')
```

```python
sns.histplot(x='price', data=df)
```

```python
sns.countplot(x='brand', data=df, order=df.brand.value_counts().iloc[:10].index).tick_params(axis='x', rotation=45)
```

```python
sns.countplot(x='type', data=df, order=df.type.value_counts().iloc[:10].index).tick_params(axis='x', rotation=45)
```
